# Clustering Image Data
Objective: To identify the dominant color in the image.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
from sklearn.cluster import KMeans
```

#### Open and display the image “dog.jpeg.”

```python
img = imread('dogs.jpeg')
plt.imshow(img)
```

```python
img
```

#### Find out the dimensions of the image and convert it to a two-dimensional array.

```python
img.shape
```

```python
doggo = img.reshape(-1, 3)
doggo.shape
```

#### Use K-means clustering with k set to 3 and cluster the image

```python
km = KMeans(n_clusters=3, random_state=42)
km.fit(doggo)
```

#### Predict the cluster label of every pixel in the image and plot it back as an image

```python
img_seg = km.cluster_centers_
img_seg
```

```python
pix_label = img_seg[km.labels_]
pix_label = pix_label.reshape(img.shape)
```

```python
pix_label
```

```python
pred = km.predict(doggo)
pred = pred.reshape((185,272))
plt.imshow(pred)
```

#### Find out the three dominant colors in the image

```python
img_seg = [x.astype(int) for x in img_seg]
img_seg
```

```python
blob = np.ones((100,100,3))
col1 = blob * img_seg[0]
plt.imshow((col1* 255).astype(np.uint8))
```

```python
blob = np.ones((100,100,3))
col2 = blob * img_seg[1]
plt.imshow((col2* 255).astype(np.uint8))
```

```python
blob = np.ones((100,100,3))
col3 = blob * img_seg[2]
plt.imshow((col3* 255).astype(np.uint8))
```
